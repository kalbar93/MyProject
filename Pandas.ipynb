```python
import pandas as pd
# ZAD 1

dane = pd.read_csv('/Users/marek.kalbarczyk/Desktop/FC/dataset_pandas_pd.csv', delimiter =',', names = ['Imię','Nazwisko','rok_urodzenia','miesiąc_urodzenia','dzień_urodzenia','miejsce_urodzenia','Miasto_oddział_firmy','departament','rok_rozpoczęcia','miesiąc_rozpoczęcia','dzień_rozpoczęcia','doświadczenie(lata)'])
```

```python
display(dane.head())
print(dane.shape)
```

```python


# ZAD 2

nazwisko = dane['Nazwisko'].value_counts()

print(nazwisko.head(1))

print(nazwisko)
```

```python
#ZAD 3

plec = []
for i in dane['Nazwisko']:
    if i[-1]== 'a':
        plec.append('K')
    else:
        plec.append('M')

print(plec)

dane.insert(2,'Płeć',plec)
```

```python
dane.head()
```

```python
#ZAD 4

dane_sort = dane.sort_values(by=['Nazwisko','rok_urodzenia'])
dane_sort.head()
```

```python
# ZAD 5
#Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób słowny, przyjmując:
#Junior (do 3 lat doświadczenia)
#Mid (4 do 8 lat)
#Senior (9 do 12 lat)
#Ekspert (od 13 lat)


def exp(i):
    if i < 3:
        return 'Junior'
    elif i > 4 & i < 8:
        return 'Mid'
    elif i >9 & i <=12:
        return 'Senior'
    else:
        return 'Ekspert'
    
dane['Ranga']=dane['doświadczenie(lata)'].apply(exp)
        
        
dane.head()
```

```python
# ZAD 6 
#Kolumna reprezentująca rok dołączenia do firmy ma puste wartości. Uzupełnij brakujące wartości używając
#średniej dla osób o takim samym poziomie doświadczenia (kolumna z Zadania 3), w skali całej firmy.

dane['rok_rozpoczęcia']=dane['rok_rozpoczęcia'].fillna(round(dane.groupby('Ranga')['rok_rozpoczęcia'].transform('mean'),0))


dane.head()
```

```python
# ZAD 7
#Który oddział firmy ma najwięcej Data Scientistów ?
dane.head()


dane[dane['departament'] == 'Data Science'].groupby('Miasto_oddział_firmy').count().idxmax().Imię
```

```python
# ZAD 8

#Szef chciałby zmienić nazewnictwo departamentów w całej firmie. Napisz kod, który zamieni wartości w
#kolumnie reprezentującej departament, nastepująco:

#Software Development -> Software Development Team
#HR -> Human Resources
#Data Science -> Data Analytics


dane = dane.replace('Software Development','Software Development Team')
dane = dane.replace('HR','Human Resources')
dane = dane.replace('Data Science','Data Analytics')

dane.head()
```

```python
# ZAD 9 
#Które biuro firmy ma najwięcej osób na Seniorskim poziomie doświadczenia ?
dane.head()
dane[dane['Ranga'] == 'Senior'].groupby('Miasto_oddział_firmy').count().idxmax().departament
```

```python
# ZAD 10

#Które biuro ma najwięcej kobiet na Juniorskim poziomie doświadczenia ?


dane[(dane['Ranga'] == 'Junior')&(dane['Płeć'] == 'K')].groupby('Miasto_oddział_firmy').count()
# brak juniorów
```

```python
# ZAD 11
#Ile jest kobiet w krakowskim biurze, na Mid poziomie doświadczenia, urodzonych przed 1975 rokiem ?

dane[(dane['Miasto_oddział_firmy']== 'Kraków') & (dane['Ranga']=='Mid') &(dane['rok_urodzenia'] <1975)].groupby('Płeć').count()
```

```python
#ZAD 12

#Ile osób, które posiadają obecnie stanowisko seniorskie, dołączyło do firmy w przeciągu ostatnich 10 lat?

dane.head()

dane[(dane['Ranga']=='Senior') & (dane['rok_rozpoczęcia'] - 10)].count()
```
